```python
import pandas as pd
data = pd.read_csv(r"C:\Users\ADITI\Desktop\Bike sharing\Bike-Sharing-Dataset\day.csv")
print(data.shape)
```

```python
#Scatter plot to show count wrt temperature
#year - 2011

import matplotlib.pyplot as plt

d1 = data[(data['mnth'] == 1) & (data['yr'] == 0)]
d2 = data[(data['mnth'] == 2) & (data['yr'] == 0)]
d3 = data[(data['mnth'] == 3) & (data['yr'] == 0)]
d4 = data[(data['mnth'] == 4) & (data['yr'] == 0)]
d5 = data[(data['mnth'] == 5) & (data['yr'] == 0)]
d6 = data[(data['mnth'] == 6) & (data['yr'] == 0)]
d7 = data[(data['mnth'] == 7) & (data['yr'] == 0)]
d8 = data[(data['mnth'] == 8) & (data['yr'] == 0)]
d9 = data[(data['mnth'] == 9) & (data['yr'] == 0)]
d10 = data[(data['mnth'] == 10) & (data['yr'] == 0)]
d11 = data[(data['mnth'] == 11) & (data['yr'] == 0)]
d12 = data[(data['mnth'] == 12) & (data['yr'] == 0)]

plt.scatter(d1['atemp'], d1['cnt'],c='r',label="Jan")
plt.scatter(d2['atemp'], d2['cnt'],c='g',label="Feb")
plt.scatter(d3['atemp'], d3['cnt'],c='b',label="Mar")
plt.scatter(d4['atemp'], d4['cnt'],c='y',label="Apr")
plt.scatter(d5['atemp'], d5['cnt'],c='c',label="May")
plt.scatter(d6['atemp'], d6['cnt'],c='m',label="Jun")
plt.scatter(d7['atemp'], d7['cnt'],c='k',label="July")
plt.scatter(d8['atemp'], d8['cnt'],c='lavender',label="Aug")
plt.scatter(d9['atemp'], d9['cnt'],c='plum',label="Sept")
plt.scatter(d10['atemp'], d10['cnt'],c='slategrey',label="Oct")
plt.scatter(d11['atemp'], d11['cnt'],c='pink',label="Nov")
plt.scatter(d12['atemp'], d12['cnt'],c='greenyellow',label="Dec")

plt.xlabel('Normalized Temperature')
plt.ylabel('Count')
plt.legend()
plt.title("Temperature vs Monthly count of users of year 2011",fontsize=15)
#plt.savefig("TempYear10"+".png",dpi=550)
plt.show()
```
