```python
#calculating monthly count of users in year 2012
import pandas as pd
dt = pd.read_csv(r"c:\Users\Sony\Desktop\day.csv")
l=[]
for i in range(1,13):
    s=0
    val = dt[(dt['mnth'] == i) & (dt['yr']==1) ]
    for i,r in val.iterrows():
     s=s+r['cnt']
    #print(s)
    l.append(s)

print(l)


```

```python
#Months vs count of user line graph
import matplotlib.pyplot as plt

month = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sept","Oct","Nov","Dec"]

#month = ["January","February","March","April","May","June","July","August","September","October","November","December"]

one=[38189, 48215, 64045, 94870, 135821, 143512, 141341, 136691, 127418, 123511, 102167, 87323]
two= [96744, 103137, 164875, 174224, 195865, 202830, 203607, 214503, 218573, 198841, 152664, 123713]

plt.plot(range(len(month)),one, color='red',label="2011",linewidth=10)
plt.plot(range(len(month)),two, color='blue',label="2012",linewidth=10)
axes = plt.gca()

axes.set_xticks(range(len(month)))
axes.set_xticklabels(month, rotation=45, ha="right",fontsize=18)
   
plt.xlabel('Months',fontsize=22)
plt.ylabel('Count of Users',fontsize=22)
plt.rcParams.update({'font.size': 18})
plt.title('Monthly count of users',fontsize=27)
plt.legend()
fig = plt.gcf() 
fig.set_size_inches(11,8)
plt.tight_layout()
plt.savefig("MonthvsCount.jpg",dpi=550)
plt.show()
```
