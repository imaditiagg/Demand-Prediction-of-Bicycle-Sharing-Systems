```python
#Boxplot

import pandas as pd
data = pd.read_csv(r"C:\Users\ADITI\Desktop\Bike sharing\Bike-Sharing-Dataset\day.csv")
print (data.shape)
```

```python
import matplotlib.pyplot as plt

#Count distribution 

plt.boxplot(data['cnt'],labels=['count'])

plt.title('Count of users')
plt.show()
```

```python
#Count distribution wrt seasons

df=data

val1 = df[(df['season']==1)]
val2 = df[(df['season']==2)]
val3 = df[(df['season']==3)]
val4 = df[(df['season']==4)]
plot_data=[val1['cnt'],val2['cnt'],val3['cnt'],val4['cnt']]
plt.boxplot(plot_data,labels=['spring','summer','fall','winter'])
plt.title('Count of users wrt seasons')

plt.show()
```

```python
#Count distribution wrt weather conditions

df=data

val1 = df[(df['weathersit']==1)]
val2 = df[(df['weathersit']==2)]
val3 = df[(df['weathersit']==3)]


plot_data=[val1['cnt'],val2['cnt'],val3['cnt']]
plt.boxplot(plot_data,labels=['Clear+Partly cloudy','Mist+Cloudy','Light Rain+Thunderstorm'])
plt.title('Count of users wrt weathersit')
plt.show()
```

```python
#Count distribution wrt holiday/No holiday

df=data
import matplotlib.pyplot as plt
val1 = df[(df['holiday']==1)]
val2 = df[(df['holiday']==0)]
plot_data=[val1['cnt'],val2['cnt']]
plt.boxplot(plot_data,labels=['Holiday','No Holiday'])
plt.title('Count of users w.r.t Holidays')
#plt.savefig("myBoxPlot"+".png",dpi=550)
plt.show()
```

```python
#count distrbution wrt weekdays

df=data

val1 = df[(df['weekday']==0)]
val2 = df[(df['weekday']==1)]
val3 = df[(df['weekday']==2)]
val4 = df[(df['weekday']==3)]
val5 = df[(df['weekday']==4)]
val6 = df[(df['weekday']==5)]
val7 = df[(df['weekday']==6)]
plot_data=[val1['cnt'],val2['cnt'],val3['cnt'],val4['cnt'],val5['cnt'],val6['cnt'],val7['cnt']]
plt.figure(figsize=(8,5))
plt.boxplot(plot_data,labels=['sunday','monday','tuesday','wednesday','thursday','friday','saturday'])
plt.title('Count of users wrt Weekdays')
plt.savefig("BoxPlot5"+".png")
```
