```python
import pandas as pd
from IPython.display import display

data = pd.read_csv(r"C:\Users\user\Desktop\College Work\sem 4\BI\project work\day.csv")
pd.options.display.max_rows = None
display(data)
```

```python
#histogram of d_temp

import matplotlib.pyplot as plt

#for month January 2011- March 2011    
data1=data.iloc[0:90, 13:14]
plt.hist(data1['d_temp'])
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Temperature Histogram for d_temp for month January 2011- March 2011')
plt.show()


#for month April 2011 - June 2011
data1=data.iloc[90:181, 13:14]
plt.hist(data1['d_temp'])
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Temperature Histogram for d_temp for month April 2011 - June 2011')
plt.show()


#for month July 2011 - September 2011
data1=data.iloc[181:274, 13:14]
plt.hist(data1['d_temp'])
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Temperature Histogram for d_temp for month July 2011 - September 2011')
plt.show()


#for month October 2011 - December 2011
data1=data.iloc[274:365, 13:14]
plt.hist(data1['d_temp'])
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.title('Temperature Histogram for d_temp for month October 2011 - December 2011')
plt.show()
```
